## 벤포드 스코어 계산

```python
import pandas as pd
import numpy as np

# 데이터 파일 경로
file_path = '/content/KAGGLE/DATASET-balanced.csv'

# 데이터 읽기
df = pd.read_csv(file_path)

# 벤포드의 법칙에 따른 예상 확률 분포
benford_dist = np.log10(1 + 1 / np.arange(1, 10))

def get_first_nonzero_digit(value):
    try:
        # 숫자로 변환
        num_value = float(value)
    except ValueError:
        return np.nan  # 변환이 안되면 NaN 반환

    # 절댓값을 취하고 문자열로 변환
    str_value = str(abs(num_value))

    # 소수점이 있는 경우
    if '.' in str_value:
        # 소수점 이후 첫 번째로 0이 아닌 숫자를 찾음
        for char in str_value:
            if char.isdigit() and char != '0':
                return int(char)
    else:
        # 정수인 경우 첫 번째 숫자 반환
        return int(str_value[0])
    return np.nan  # 예외적으로 처리할 값이 없는 경우 NaN 반환

# 각 행마다 벤포드의 첫 번째 숫자 분포를 확인하는 함수
def calculate_benford_row(row):
    first_digits = [get_first_nonzero_digit(value) for value in row]
    first_digits = pd.Series(first_digits).dropna().astype(int)

    # 각 숫자의 빈도 계산
    digit_counts = first_digits.value_counts().sort_index()

    # 관측된 첫 번째 숫자의 비율 계산
    observed_dist = digit_counts / digit_counts.sum()

    # 기대되는 벤포드 분포와의 차이 계산
    expected_counts = benford_dist * digit_counts.sum()
    observed_counts = digit_counts.reindex(np.arange(1, 10), fill_value=0)

    # 카이제곱 검정으로 점수 계산
    chi2_stat = ((observed_counts - expected_counts) ** 2 / expected_counts).sum()
    return chi2_stat

# 라벨 컬럼은 따로 저장
label_column = df['LABEL']

# 각 행에 대해 벤포드 점수를 계산하여 새로운 피처로 추가
df_benford = df.drop(columns='LABEL').apply(calculate_benford_row, axis=1)
df['benford_score'] = df_benford


# 벤포드 피처가 추가된 데이터프레임 확인
print(df.head())

# 벤포드 피처를 추가한 CSV 파일 저장
output_path = '/content/KAGGLE/DATASET-balanced-benford.csv'
df.to_csv(output_path, index=False)
print(f"벤포드 피처가 추가된 파일이 저장되었습니다: {output_path}")
```

# LSTM

## 데이터 전처리

```python
def prepare_data(X, window_size = 10):

    data = []

    for i in range(len(X)):
        row = X.iloc[i].values
        row_data = []
        for j in range(len(row) - window_size):
            window = row[j : j + window_size]
            row_data.append(window)
        data.append(row_data)

    return np.array(data)
```

```python
X.iloc[:, :] = MinMaxScaler().fit_transform(X)
new_X = prepare_data(X, window_size = 5)
```

## 데이터 분리(트레이닝, 테스트)

```python
X_train, X_test, y_train, y_test = train_test_split(new_X, y.values, test_size=0.2, shuffle = True, stratify = y.values, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, shuffle = True, stratify = y_train, random_state=42)
```

## LSTM 모델 설계

```python
# 모델 초기화
model = Sequential()

input_shape = (X_train.shape[1], X_train.shape[2])
model.add(Input(shape=input_shape))
model.add(Bidirectional(LSTM(128, return_sequences=True)))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Bidirectional(LSTM(128, return_sequences=True)))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Bidirectional(LSTM(64, return_sequences=False)))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

optimizer = optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss=BinaryCrossentropy(), metrics=['accuracy'])
model.summary()
```

모델 체크포인트 저장 선언

```python
callback = [
    ModelCheckpoint(filepath='kaggle_model.keras'),
]
```

## LSTM 모델 학습

```python
history = model.fit(X_train, y_train, batch_size=32, epochs=60, validation_data = (X_val, y_val), callbacks = callback)
```

## 추가 데이터 테스트

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import load_model
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, recall_score, precision_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder

# 파일 경로 설정
vocal_test_path = '/content/add_kaggle_test_benford.csv'
model_path = '/content/kaggle_model.keras'

# 데이터 로드
vocal_test = pd.read_csv(vocal_test_path)

# 라벨 컬럼 분리
y_vocal_test = vocal_test['LABEL']
X_vocal_test = vocal_test.drop(columns=['LABEL'])

# 같은 LabelEncoder를 사용하여 라벨 인코딩
label_encoder = LabelEncoder()
label_encoder.fit(['REAL', 'FAKE'])  # 학습 데이터의 라벨 값으로 인코딩 학습
y_vocal_test_encoded = label_encoder.transform(y_vocal_test)

# 모델 로드
model = load_model(model_path)

# 데이터의 형태를 모델에 맞게 변환 (입력 데이터는 (샘플 수, 21, 5) 형식이어야 함)
def reshape_data(data, window_size=5):
    reshaped_data = []
    for i in range(len(data)):
        row = data.iloc[i].values
        reshaped_row = []
        for j in range(len(row) - window_size + 1):
            window = row[j: j + window_size]
            reshaped_row.append(window)
        reshaped_data.append(reshaped_row)
    return np.array(reshaped_data)

# 데이터 리셰이프
X_vocal_test_reshaped = reshape_data(X_vocal_test)

# 예측 수행
y_vocal_pred = np.round(model.predict(X_vocal_test_reshaped).flatten())

# 성능 지표 계산 함수
def evaluate_performance(y_true, y_pred, dataset_name):
    accuracy = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    roc_auc = roc_auc_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    conf_matrix = confusion_matrix(y_true, y_pred)

    print(f"Performance on {dataset_name}:")
    print(f"Accuracy: {accuracy}")
    print(f"F1 Score: {f1}")
    print(f"ROC-AUC Score: {roc_auc}")
    print(f"Recall: {recall}")
    print(f"Precision: {precision}")
    print("\n")

    # Confusion matrix 시각화
    plt.figure(figsize=(8, 6))
    ax = sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
    ax.set_xlabel('Predicted Labels')
    ax.set_ylabel('True Labels')
    ax.set_title(f'Confusion Matrix for {dataset_name}')
    ax.set_xticklabels(['FAKE', 'REAL'])
    ax.set_yticklabels(['FAKE', 'REAL'])
    plt.show()

# 성능 평가 및 confusion matrix 시각화
evaluate_performance(y_vocal_test_encoded, y_vocal_pred, "Korean Test Dataset")
```
